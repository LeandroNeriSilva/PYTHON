```python
from random import randint
```

```python
def Jogo_numero_da_sorte():
  fichas = int(input('Informe o numero de fichas que deseja jogar:\n= '))
  while fichas >= 1:
    numero_da_sorte = int(input('Informe seu numero da sorte:\n= '))
    numero_sorteado = randint(0,5)
    if numero_da_sorte == numero_sorteado and fichas > 0:
      fichas = fichas + 1
      print('GANHOOOU!!! Voce possui {} fichas, continue jogando!!!'.format(fichas))
    elif numero_da_sorte != numero_sorteado and fichas > 1:
      fichas = fichas - 1
      print('Tente Novamente!!! Voce possui {} fichas, continue jogando!!!'.format(fichas))
    else:
      fichas = fichas - 1
      print('Você errou sua ultima ficha, suas chances acabaram, compre mais fichas!!!')
```

```python
Jogo_numero_da_sorte()
```
