```python
# importando biblioteca
import pandas as pd
```

```python
# capturamos a url de interesse e aplicamos ',' para os numeros decimais e '.' para numeros em milhares
# sem esse parametro os numeros seriam exibidos sem separadores 
url = 'https://pt.wikipedia.org/wiki/Salvador'
site = pd.read_html(url, decimal=',', thousands='.')
```

```python
# vamos filtrar o indice de interesse, neste indice contem a tabela que desejamos
site[4]
```

```python
# vamos atribuir a uma variavel que vamos chamar de 'salvador'
salvador = site[4]
```

```python
# vamos conhecer os indices para aplicarmos uma condição
print(salvador.columns[0][1])
print(salvador.columns[1][1])
print(salvador.columns[2][1])
```

```python
# começamos criando uma lista vazia que vamos chamar de 'cols'
# para cada 'i' item em um range que compreende o 'len' tamanho de salvador.columns
# pegue cada 'i' item e acrescente em cols conforme (salvador.columns[i][1])
cols = []
for i in range(len(salvador.columns)):
  cols.append(salvador.columns[i][1])
cols[0] = 'Medidas'
```

```python
# atribuimos a salvador.columns o resultado de cols 
salvador.columns = cols
```

```python
# excluimos a linha de indice 9
salvador.drop(9, inplace=True)
```

```python
# verificando
salvador
```

```python
# vamos atribuir 'Medidas' como indice
salvador.index = salvador['Medidas']
```

```python
# agora temos 'Medidas' duas vezes em nosso dataframe, vamos remove um
salvador.drop('Medidas', axis=1, inplace=True)
```

```python
# verificando
salvador
```

```python
# vamos excluir a ultima coluna 'Ano'
salvador.drop('Ano', axis=1, inplace=True)
```

```python
# verificando
salvador
```

```python
# vamos alterar os nomes das colunas
salvador.rename(
    columns={'Jan':'Janeiro', 'Fev':'Fevereiro', 'Mar':'Março', 'Abr':'Abril', 'Mai':'Maio', 'Jun':'Junho',
             'Jul':'Julho', 'Ago':'Agosto', 'Set':'Setembro', 'Out':'Outubro', 'Nov':'Novembro', 'Dez':'Dezembro'},
             inplace=True
)
```

```python
# verificando
salvador
```
